#**Titanic**

```python

datafile = pd.read_csv('titanic.csv')
print(datafile)
```

```python
datafile[['Name','Sex']] = datafile[['Name','Sex']].astype('category')
print(datafile.dtypes)
```

```python
correlation = datafile.corr()
sns.set(rc = {'figure.figsize':(10,10)})
sns.heatmap(correlation,annot = True)
```

```python
sns.boxplot(x = datafile["Age"]);
plt.show()
```

```python
data = datafile['Sex']
sns.displot(data)
plt.show()
```

```python
plt.figure(figsize=(20,20))
datafile['Survived'].value_counts().plot(kind='pie',labels =['Not survived','Survived'], autopct='%1.1f%%')
plt.legend(labels =['Not survived','Survived'])
plt.title("Percentage of people that survived vs not survived the tragedy")
```

```python
sns.barplot(x = 'Pclass',y = 'Survived',data = datafile)
```

```python
sns.barplot(x = 'Parents/Children Aboard',y = 'Survived',data = datafile)
```

#**Netflix**


```python

nfdf = pd.read_csv('netflix_titles.csv')
print(nfdf)
```

```python
nfdf.describe()
```

```python
nfdf.tail()
```

```python
nfdf.nunique(axis=0)
```

```python
nfdf.keys()
```

```python
nfdf.dtypes
```

```python
nfdf.isnull().sum(axis=0)
```

```python
nfdf.dropna(inplace= True)
nfdf.isnull().sum(axis=0)
```

```python
listnew = list(nfdf.keys())
del listnew[7]
print(listnew)
nfdf[listnew] = nfdf[listnew].astype('category')
print(nfdf.dtypes)
```

```python
sns.displot(data=nfdf, x="rating", col="type",height=10, aspect=1.2)
plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(15, 12)
sns.scatterplot(data = nfdf, x = "release_year", y = "rating")

plt.show()
```

```python
nfdf['type'].value_counts().plot(kind='barh')
```

```python
plt.figure(figsize=(12,12))
plt.hist(nfdf.release_year)
plt.xlabel("Movies/TV shows release year")
plt.ylabel("No. of Movies/TV Shows released that year")
```

```python
plt.figure(figsize=(20,20))
nfdf['country'][:10].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Percentage of movies/shows released in top 10 countries")
```
